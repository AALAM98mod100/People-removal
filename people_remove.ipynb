```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

image_stack = [cv.imread(f"./database/ams/{i}.jpg") for i in range(1,21)]

# cv.imshow("Image", image_stack[0])
plt.imshow(image_stack[0])
```

```python
def median_array(array):
    return np.median(array,axis=0)

height,width,_ = image_stack[0].shape
```

```python
for i in range(len(image_stack)):
    image_stack[i] = cv.cvtColor(image_stack[i], cv.COLOR_BGR2HSV)
```

```python
result = np.zeros((height,width,3), np.uint8)
for i in range(height):
    for j in range(width):
        result[i,j] = median_array([image_stack[k][i,j] for k in range(len(image_stack))])

plt.imshow(cv.cvtColor(result, cv.COLOR_HSV2RGB))
```
